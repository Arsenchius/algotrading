```python
from binance import Client
import pandas as pd
import pandas_ta as pta
```

```python
INTERVAL = "1h"
symbol = "BTCUSDT"
end_date = "20000 hours ago UTC"
```

```python
def get_actual_data(
    symbol: str,
    time_frame: str = INTERVAL,
    end_date: str = "20000 hours ago UTC",
) -> pd.DataFrame:
    client = Client()
    df = pd.DataFrame(client.get_historical_klines(symbol, time_frame, end_date))

    df = df.iloc[:, 0:6]
    df.columns = ["Time", "Open", "High", "Low", "Close", "Volume"]
    df.set_index("Time", inplace=True)
    df.index = pd.to_datetime(df.index, unit="ms")
    df = df.astype(float)
    return df
```

```python
data = get_actual_data(symbol, end_date = end_date)
data['RSI'] = pta.rsi(data['Close'], length=21)
data
```
